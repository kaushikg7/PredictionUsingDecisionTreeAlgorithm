```python
from sklearn import datasets
import pandas as pd
iris=datasets.load_iris()
dir(iris)
print(iris.DESCR)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print(df)
```

```python
df['target']=pd.Series(iris.target)
print(df)
```

```python
df.sample(n=6)
```

```python
x=df['target_names']=df['target'].apply(lambda y:iris.target_names[y])
print(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test=train_test_split(df,test_size=0.3)
```

```python
print("number of training samples")
df_train.shape[0]
```

```python
print("number of testing samples")
df_test.shape[0]
```

```python
from sklearn.tree import export_text
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
x_train=df_train[iris.feature_names]
x_test=df_test[iris.feature_names]
```

```python
y_train=df_train['target']
y_test=df_test['target']
```

```python
clf.fit(x_train,y_train)
```

```python
y_test_pred=clf.predict(x_test)
```

```python
print("class of testing samples")
print(y_test_pred)
```

```python
text_representation =tree.export_text(clf)
```

```python
from sklearn.tree import export_text
from sklearn import tree
from matplotlib import pyplot as plt
print(text_representation)
```

```python
with open("decision_tree.log","w") as foot :
  foot.write(text_representation)
fig=plt.figure(figsize=(25,20))
_=tree.plot_tree(clf,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
fig.savefig("decision_tree.png")
```

```python
from sklearn.metrics import accuracy_score
x=accuracy_score(y_test,y_test_pred)
print("accuracy")
print(x)
```
